## Overall Employee Satisfaction
JobSatisfaction column contains employee ratings from 1 to 4. To get an overall sense of employee satisfaction, the approach will be as follows:
1. Create a dictionary where:
    * 1 => Low
    * 2 => Medium
    * 3 => High
    * 4 => Very High
2. Use the map method to transform the code into the four categories, such that 1 will turn into 'Low', 2 into 'Medium', etc.
3. Reassign the Series

```python
df['JobSatisfaction'].head()
```

```python
# Create the category map
category_map = {
    1: "Low",
    2: "Medium",
    3: "High",
    4: "Very High"
}
```

```python
# Apply the category dictionary using Series' map method
# to transform the encodings into categorical labels
# and reassign the Series
df['JobSatisfaction'] = df['JobSatisfaction'].map(category_map)
df['JobSatisfaction'].head()
```

```python
df['JobSatisfaction'].head()
```

```python
# Get the frequency of each category occurrence using
# the value_counts method

### Your Code Below ###

df['JobSatisfaction'].value_counts(normalize=True)
```

```python
# To get a sense of the overall satisfaction, should
# look at the proportion. Multiply by 100 to get %

### Your Code Below ###
df['JobSatisfaction'].value_counts(normalize=True)*100

```

Let's compare employees with Low and Very High JobSatisfaction levels based on:
* `Age`
* `Department`
* `DistanceFromeHome`
Make the comparisons by grouping first, then apply the sequence of actions: split, apply, combine.

Steps in grouping operations:
1. **Split** DataFrame into group of DataFrames based on required criteria or categorical variable. The results of a pandas grouped object is similar to a dictionary => every group is associated with a different key
2. **Apply** a function to every part of this grouped object
3. **Group/Combine** the results of the previous operation into a new data structure.

To make the comparisons:
1. Create a new DataFrame containing only observations that meet the criteria for comparison and call it `subset_of_interest`
2. Use the `bitwise` or `|` operator to get Series of employees with Low and High `JobSatisfaction` and combine them and use the `shape` attribute to see the shape of the newly created DataFrame
3. Then use `value_count` to see exact counts of each observation.

```python
subset_of_employees = df.loc[(df['JobSatisfaction'] == 'Low') | 
                              (df['JobSatisfaction'] == 'Very High')]
subset_of_employees.shape                                       
```

```python
# get the counts of this subset using the value_counts method

### Your Code Below ###
subset_of_employees['JobSatisfaction'].value_counts()

```

```python
# get the proportion by passing the normalize=True argument

### Your Code Below ###

subset_of_employees['JobSatisfaction'].value_counts(normalize=True)*100
```

We now have the group of observations we're interested in. Apply `groupby` operation to the new subset to compare across the variables we're interested in.

Use `get_group` method to get the data associated with each group.

```python
# group this subset of employees by "JobSatisfaction" using the .groupby then use get_group on the groupd object to select the "Low" group

### Your Code Below ###
grouped = subset_of_employees.groupby("JobSatisfaction")
grouplow = grouped.get_group("Low")
grouplow
```

```python
# find the average "Age" of this group

### Your Code Below ###
grouped["Age"].mean()

```

```python
# use describe on the "Age" group for more stats

### Your Code Below ###

grouped["Age"].describe()
```

```python
# select both "Low" and "Age" and describe the stats from the group

### Your Code Below ###
grouplow["Age"].describe()

```

```python
# Do the same but for "Very High" and "Age"

### Your Code Below ###
grouphigh = grouped.get_group("Very High")
grouphigh["Age"].mean()

```

```python
# Now do the same but look at other columns such as 'DistanceFromHome',
# 'Gender', and compare to "Low" or "Very High" groups

### Your Code Below ###
grouplow["DistanceFromHome"].describe()

```

```python
grouphigh["DistanceFromHome"].describe()
```
